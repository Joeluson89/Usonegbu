```python
# Import packages

#required for manipulating data
import pandas as pd
import numpy as np

#required for building the interactive dashboard
import panel as pn
pn.extension('tabulator')
import hvplot.pandas
import holoviews as hv
hv.extension('bokeh')

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Revolut.csv')
df
```

```python
a = df['Description'].unique()
print(a)
```

```python
# Clean the data
# Keeping only desired columns for the data analysis
df = df[['Completed Date', 'Description', 'Amount']]
df['Description'] = df['Description'].map(str.lower)

# Renaming a column from Completed Date to Date
df = df.rename(columns={'Completed Date' : 'Date'})

# Input a new column called Category and the values are 'Unassigned'
df['Category'] = 'unassigned'

df.head()
```

```python
# Define categories

    # Groceries
    # Shopping
    # Transport
    # Gifts

 # Groceries
df['Category'] = np.where(df['Description'].str.contains(
    "pharmacy|o2|deliveroo|kokoro|sainsbury's|klarna|tesco|greggs|superdrug|wilko|morrisons|supersave wood green"),
"Groceries", df['Category'])

# Gifts
df['Category'] = np.where(df['Description'].str.contains("cupcakes|food palace|from jolomi samuel mogbeyiteren|payment from farouk zekeri"),
"Gifts", df['Category'] )

# Payments/Transfers
df['Category'] = np.where(df['Description'].str.contains("payment from springcare support|to ekene joel usonegbu|payment from ekene usonegbu"),
"Payments/Transfers", df['Category'] )

# Convert the Date Column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract the month and year information
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

pd.options.display.max_rows = 999
df.head(10)
```
